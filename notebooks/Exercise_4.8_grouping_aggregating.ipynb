### Step 2: Mean order number by department (entire dataset)

```python
# --------------------------------------------------
# Step 2: Aggregate mean of order_number by department_id
# (Using entire dataset, not a subset)
# --------------------------------------------------

dept_order_mean = ords_prods_merge.groupby('department_id').agg({
    'order_number': ['mean']
})

dept_order_mean
```

### Step 3: Comparison of full dataset vs subset results

When comparing the mean order numbers by department calculated on the full dataset to those calculated earlier on the 1,000,000-row subset, the overall patterns remain consistent across departments. Departments with higher average order numbers in the subset analysis also show higher averages in the full dataset.

However, small differences in the mean values are observed. These differences are expected because the subset represents only a sample of the total data, while the full dataset includes all available orders and customers. Using the full dataset reduces sampling bias and provides more stable and representative averages.

Overall, the full dataset results confirm the trends observed in the subset while offering a more accurate reflection of customer ordering behavior across Instacart departments.

```python
# ------------------------------------------------------------
# Step 4A: Create max_order column using transform()
# (Maximum order number per user)
# ------------------------------------------------------------

ords_prods_merge['max_order'] = (
    ords_prods_merge
    .groupby('user_id')['order_number']
    .transform('max')
)
```

```python
# ------------------------------------------------------------
# Step 4B: Validate max_order column
# ------------------------------------------------------------

ords_prods_merge[['user_id', 'order_number', 'max_order']].head(20)
```

```python
# ------------------------------------------------------------
# Step 4C: Create loyalty_flag based on max_order
# ------------------------------------------------------------

# Initialize column
ords_prods_merge['loyalty_flag'] = 'Not classified'

# Loyal customers (> 40 orders)
ords_prods_merge.loc[
    ords_prods_merge['max_order'] > 40,
    'loyalty_flag'
] = 'Loyal customer'

# Regular customers (11–40 orders)
ords_prods_merge.loc[
    (ords_prods_merge['max_order'] > 10) & (ords_prods_merge['max_order'] <= 40),
    'loyalty_flag'
] = 'Regular customer'

# New customers (10 orders or fewer)
ords_prods_merge.loc[
    ords_prods_merge['max_order'] <= 10,
    'loyalty_flag'
] = 'New customer'
```

```python
# ------------------------------------------------------------
# Step 4D: Check loyalty flag distribution
# ------------------------------------------------------------

ords_prods_merge['loyalty_flag'].value_counts()
```

```python
# ------------------------------------------------
# Step 5A: Analyze product prices by loyalty group
# ------------------------------------------------

price_by_loyalty = ords_prods_merge.groupby('loyalty_flag').agg({
    'prices': ['mean', 'min', 'max']
})

price_by_loyalty
```

### Step 5 – Price Differences by Loyalty Group

The average product price differs across loyalty segments. New customers have the highest mean product price, followed by regular customers, while loyal customers have the lowest average price. This suggests that newer users tend to purchase more expensive items, whereas loyal customers may focus on frequently purchased or lower-priced staple products. These findings indicate that spending behavior varies by customer loyalty level.

```python
# --------------------------------------------------
# Step 6A: Calculate mean product price per user
# (Average price across all orders for each user)
# --------------------------------------------------

ords_prods_merge['mean_price'] = (
    ords_prods_merge
    .groupby('user_id')['prices']
    .transform('mean')
)
```

```python
# --------------------------------------------------
# Step 6B: Validate mean_price column
# --------------------------------------------------

ords_prods_merge[['user_id', 'prices', 'mean_price']].head(20)
```

```python
# --------------------------------------------------
# Step 6C: Create spending_flag based on mean_price
# --------------------------------------------------

# Initialize column
ords_prods_merge['spending_flag'] = 'Not classified'

# Low spenders (mean price < 10)
ords_prods_merge.loc[
    ords_prods_merge['mean_price'] < 10,
    'spending_flag'
] = 'Low spender'

# High spenders (mean price >= 10)
ords_prods_merge.loc[
    ords_prods_merge['mean_price'] >= 10,
    'spending_flag'
] = 'High spender'
```

```python
# --------------------------------------------------
# Step 6D: Check spending_flag distribution
# --------------------------------------------------

ords_prods_merge['spending_flag'].value_counts()
```

```python
# --------------------------------------------------
# Step 7A: Calculate median days_since_prior_order per user
# --------------------------------------------------

ords_prods_merge['median_days_between_orders'] = (
    ords_prods_merge
    .groupby('user_id')['days_since_prior_order']
    .transform('median')
)
```

```python
# --------------------------------------------------
# Step 7B: Validate median_days_between_orders column
# --------------------------------------------------

ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_between_orders']].head(20)
```

```python
# ------------------------------------------------
# Step 7C: Create order_frequency_flag
# (Based on median_days_between_orders)
# ------------------------------------------------

# Initialize column
ords_prods_merge['order_frequency_flag'] = 'Not classified'

# Non-frequent customers (median > 20 days)
ords_prods_merge.loc[
    ords_prods_merge['median_days_between_orders'] > 20,
    'order_frequency_flag'
] = 'Non-frequent customer'

# Regular customers (median > 10 and <= 20 days)
ords_prods_merge.loc[
    (ords_prods_merge['median_days_between_orders'] > 10) &
    (ords_prods_merge['median_days_between_orders'] <= 20),
    'order_frequency_flag'
] = 'Regular customer'

# Frequent customers (median <= 10 days)
ords_prods_merge.loc[
    ords_prods_merge['median_days_between_orders'] <= 10,
    'order_frequency_flag'
] = 'Frequent customer'
```

```python
# ------------------------------------------------
# Step 7D: Check order frequency distribution
# ------------------------------------------------

ords_prods_merge['order_frequency_flag'].value_counts()
```
