```python
import pandas as pd
import numpy as np
import os
```

```python
# Import cleaned orders dataset (from Prepared Data)
df_ords = pd.read_csv(
    os.path.join(
        PROJECT_ROOT,
        "02 Data",
        "Prepared Data",
        "orders_checked.csv"
    )
)
```

```python
df_ords.shape
df_ords.head()
```

```python
# 4.6 - Import combined prior orders/products dataset (from Original Data)
df_ords_prior = pd.read_csv(
    ORIGINAL_DATA / "4.6_orders_products_prior" / "orders_products_prior.csv"
)

df_ords_prior.shape
df_ords_prior.head()
```

```python
df_ords_prior.shape
```

```python
df_ords.shape
```

```python
df_ords_products = pd.merge(
    df_ords,
    df_ords_prior,
    on='order_id',
    how='left'
)
```

```python
df_ords_products.shape
df_ords_products.head()
```

```python
df_ords_products['order_id'].isna().sum()
df_ords_products['product_id'].isna().sum()
```

```python
df_ords_products.to_pickle(
    os.path.join(
        PROJECT_ROOT,
        "02 Data",
        "Prepared Data",
        "orders_products_combined.pkl"
    )
)
```
